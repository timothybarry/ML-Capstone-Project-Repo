```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("dataset_24_mushroom.csv") 
data.head()
```

```python
# looking for missing data denoted by '?'

for col in data:
    print(col + ": " + str(data[data[col] == "?"][col].count()))
```

```python
# Remove all rows with missing data in 'stalk-root column

data_row_clean = data[data["stalk-root"] != '?']

for col in data:
    print(col + ": " + str(data_row_clean[data_row_clean[col] == "?"][col].count()))
    
print("number of rows: " + str(data_row_clean.shape[0]))

data_row_clean.describe()
```

```python
# Remove missing data by dropping column with missing data

data_column_clean = data.drop("stalk-root", axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data_row_clean['class'] = le.fit_transform(data_row_clean['class'])
data_column_clean['class'] = le.fit_transform(data_column_clean['class'])
data_row_clean.head()
```

```python
# Encode categorical features

data_row_encoded = pd.get_dummies(data_row_clean)
data_column_encoded = pd.get_dummies(data_column_clean)
data_row_encoded.head(5)
# data_row_encoded.corr
```

```python
X = data_row_encoded.drop('class', axis=1)
y = data_row_encoded['class'].values.reshape(-1, 1)
type(X)
corr = X.corr()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# sc = StandardScaler()

# X = sc.fit_transform(X)

# pca = PCA(n_components = 98)

# X = pca.fit_transform(X)

# explained_variance = np.cumsum(pca.explained_variance_ratio_)

# plt.plot(explained_variance)

# plt.title('PCA:Cumulative Explained Variance')

# plt.xlabel('Components')

# plt.ylabel('Cumulative  SUM')

# pd.DataFrame(data=X[1:,1:], index=X[1:,0], columns=X[0,1:]) 

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=123, stratify=y)
```

```python
logreg = LogisticRegression()

logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
print (accuracy_score(y_test, y_pred))
type(X_train)
```

```python
coefs = pd.Series(logreg.coef_[0], index=X_train.columns)
print( X_train.columns)
coefs = coefs.sort_values()
plt.subplot(1,1,1)
coefs.plot(kind="bar")
plt.show()
print (coefs.sort_values(ascending = False))
```
